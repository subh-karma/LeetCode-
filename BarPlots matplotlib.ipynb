```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression

# Generate sample data
np.random.seed(0)
x = np.random.rand(100, 1) * 10
y = np.random.rand(100, 1) * 10
z = 2 + 3 * x + 4 * y + np.random.randn(100, 1) * 2

# Prepare the data for linear regression
X = np.hstack((x, y))

# Fit the linear regression model
model = LinearRegression()
model.fit(X, z)

# Create a mesh grid for plotting the regression plane
x_surf, y_surf = np.meshgrid(np.linspace(0, 10, 100), np.linspace(0, 10, 100))
z_surf = model.intercept_ + model.coef_[0][0] * x_surf + model.coef_[0][1] * y_surf

# Plot the data and the regression plane
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the original data points
ax.scatter(x, y, z, color='blue', label='Data points')

# Plot the regression plane
ax.plot_surface(x_surf, y_surf, z_surf, color='red', alpha=0.5, label='Regression plane')

# Set labels
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Linear Regression')

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans

# Generate sample data
np.random.seed(0)
x = np.random.rand(100, 1) * 10
y = np.random.rand(100, 1) * 10
z = np.random.rand(100, 1) * 10
data = np.hstack((x, y, z))

# Fit the K-means model
kmeans = KMeans(n_clusters=3)
kmeans.fit(data)
labels = kmeans.labels_
centers = kmeans.cluster_centers_

# Plot the data and cluster centers
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the original data points with color based on cluster assignment
ax.scatter(x, y, z, c=labels, cmap='viridis', marker='o')

# Plot the cluster centers
ax.scatter(centers[:, 0], centers[:, 1], centers[:, 2], c='red', s=200, marker='x', label='Centroids')

# Set labels
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D K-means Clustering')

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import make_classification
from sklearn.tree import DecisionTreeClassifier

# Generate sample data
np.random.seed(0)
X, y = make_classification(n_samples=100, n_features=3, n_informative=3, n_redundant=0, n_clusters_per_class=1)

# Fit the decision tree model
clf = DecisionTreeClassifier()
clf.fit(X, y)

# Create a mesh grid for plotting decision boundaries
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
z_min, z_max = X[:, 2].min() - 1, X[:, 2].max() + 1
xx, yy, zz = np.meshgrid(np.arange(x_min, x_max, 0.1),
                         np.arange(y_min, y_max, 0.1),
                         np.arange(z_min, z_max, 0.1))

# Predict class for each point in the mesh grid
grid_points = np.c_[xx.ravel(), yy.ravel(), zz.ravel()]
Z = clf.predict(grid_points)
Z = Z.reshape(xx.shape)

# Plot the data points and decision boundaries
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the original data points
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap='viridis', marker='o', label='Data points')

# Plot the decision boundaries (3D isosurfaces)
ax.contourf(xx, yy, zz, Z, alpha=0.3, cmap='viridis')

# Set labels
ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.set_zlabel('Feature 3')
ax.set_title('3D Decision Tree Classification')

plt.legend()
plt.show()
```
